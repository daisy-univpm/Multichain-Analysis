```python
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
from sklearn.linear_model import LinearRegression
sns.set_theme(style='ticks')
```

```python
path2_01 = pd.read_csv('longest_pathsv2/path_L2_H0.1.csv')['Blockchain_1']
path2_05 = pd.read_csv('longest_pathsv2/path_L2_H0.5.csv')['Blockchain_1']
path2_1 = pd.read_csv('longest_pathsv2/path_L2_H1.csv')['Blockchain_1']
```

```python
SMC2_01 = path2_01[path2_01=='BNB CHAIN'].count()
AVA2_01 = path2_01[path2_01=='Avalanche'].count()
POL2_01 = path2_01[path2_01=='Polygon'].count()
ETH2_01 = path2_01[path2_01=='BNB CHAIN'].count()
FTM2_01 = path2_01[path2_01=='Fantom'].count()

SMC2_05 = path2_05[path2_05=='BNB CHAIN'].count()
AVA2_05 = path2_05[path2_05=='Avalanche'].count()
POL2_05 = path2_05[path2_05=='Polygon'].count()
ETH2_05 = path2_05[path2_05=='BNB CHAIN'].count()
FTM2_05 = path2_05[path2_05=='Fantom'].count()

SMC2_1 = path2_1[path2_1=='BNB CHAIN'].count()
AVA2_1 = path2_1[path2_1=='Avalanche'].count()
POL2_1 = path2_1[path2_1=='Polygon'].count()
ETH2_1 = path2_1[path2_1=='BNB CHAIN'].count()
FTM2_1 = path2_1[path2_1=='Fantom'].count()
```

```python
labels = ['Avalanche','Ethereum','Fantom','Polygon','Smart Chain']
len01 = pd.Series([AVA2_01, ETH2_01, FTM2_01, POL2_01, SMC2_01]) 
len05 = pd.Series([AVA2_05, ETH2_05, FTM2_05, POL2_05, SMC2_05])
len1 =  pd.Series([AVA2_1, ETH2_1, FTM2_1, POL2_1, SMC2_1])
x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars
fig, ax = plt.subplots(figsize=(8, 5))

r1 = np.arange(len(labels))
r2 = [x + width for x in r1]
r3 = [x + width for x in r2]

ax.bar(r1, len01*100/(len01.sum()), width, label='0.1h')
ax.bar(r2, len05*100/(len05.sum()), width, label='0.5h')
ax.bar(r3, len1*100/(len1.sum()), width, label='1.0h')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel( 'Percentage of paths',**{'fontsize':12})
ax.set_xlabel('Source blockchain', **{'fontsize':12})
plt.xticks([r + width for r in range(len(labels))], labels)
ax.legend()
plt.ylim(0,35)
plt.grid()

plt.show()
fig.savefig('imgs/Grafici/percentuale-path-partenza-lunghezza-2.pdf', dpi=100, bbox_inches='tight')
```
